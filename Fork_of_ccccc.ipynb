```python

import torch_xla.core.xla_model as xm
import torch_xla.distributed.xla_multiprocessing as xmp
# Training function
epoch = 370
def train_model(arg, config,):
    print(arg)
    from torch.nn.utils.rnn import pad_sequence
    import torch as pt
    import torch
    import torcheval
    import torch.nn as nn
    from torch.utils.data import TensorDataset, DataLoader
    import pandas as pd
    import matplotlib.pyplot as plt
    import torch.nn.functional as F
    import pprint
    import pandas as pd
    import numpy as np
    import wandb
    import torch_xla
    import torch_xla.core.xla_model as xm
    import torch_xla.distributed.data_parallel as dp
    import torch_xla.distributed.xla_multiprocessing as xmp


    api_key = 'd05e1e07a877fe8e5ef5076a26698c05ec6b62ac'
    wandb.login(key=api_key)

    sweep_config = {
         'method': 'bayes',  # random, grid or bayes
        'name': 'sweep-bayes-10-200',
        'metric': {'goal': 'minimize', 'name': 'edit_distance'},
        }
    parameters_dict = {

        # 'fc_layer_size': {
        #     'values': [128, 256, 512]
        #     },
        'num_heads': {
            'values': [1,4]
        },
        'char_embed_dim': {
            'values': [32,8,128]
        },
        'dropout': {
              'values': [ 0.3]
        },
        'embed_dim': {
            'values': [4, 8,16]
        },
        'fc_dim': {
            'values': [256,512, 128]

        },
        'batch_size': {
            'values': [8,16,256]
        },
        'learning_rate': {
            'values': [0.0001]
        },
        'num_layers': {
            'values': [1,2,4]
        },

        'split_method': {
            'values': [1]
        },
        'auxilary_present': {
            'values': [0,1,-1] #no, yes, auxilary as output
        },
        'scheduler_tmax': {
            'values': [10,40,4000]
        }


    }
    config = {
    'dropout': 0.3,
    'fc_dim': 256,
    'embed_dim': 8,
    'num_layers': 8,
    'num_heads': 16,
    'char_embed_dim': 8,
    'learning_rate': 0.00015,
    'batch_size': 512,
    'split_method': 1,
    'auxilary_present': 1,
    "scheduler_tmax":4000,
   }

    sweep_config['parameters'] = parameters_dict
    # sweep_id = wandb.sweep(sweep_config, project="Character-Level Text genration")
    # print(sweep_id)
    sweep_id = "qdzgaxuz"
    text_table = wandb.Table(columns=["run","epoch", "loss", "text"])



    pprint.pprint(sweep_config)



    # Define the ALPHABET and padding value
    ALPHABET = ["<PAD>"] + list(' abcdefghijklmnopqrstuvwxyz')
    PADDING_VALUE = 0

    # Calculate the length of the ALPHABET
    ALPHABET_LENGTH = len(ALPHABET)

    # Specify the device


    from datasets import load_dataset

    # Load the TinyStories dataset
    dataset = load_dataset("roneneldan/TinyStories")
    train_data = dataset['train']

    # Extract the first 200 characters of each text entry, limiting to 2000 entries
    processed_texts = [text[:200] for text in train_data['text']][:2000]

    # Clean and filter the text based on the defined ALPHABET
    filtered_texts = [
        "".join([char for char in text.lower() if char in ALPHABET])
        for text in processed_texts
    ]

    # Print the first 40 cleaned texts
    print(filtered_texts[:100])

    # Calculate the maximum length of valid substrings across all filtered texts
    max_valid_length = max(
        len(text[:index]) for text in filtered_texts for index in range(1, len(text) + 1)
    )
    MAX_SEQUENCE_LENGTH = max_valid_length

    # Print the maximum sequence length
    print(MAX_SEQUENCE_LENGTH)


# import torch_xla
# import torch_xla.core.xla_model as xm
#d05e1e07a877fe8e5ef5076a26698c05ec6b62ac
    import torch as pt
    from torch.nn.utils.rnn import pad_sequence

    def encode(text, padding_value=PADDING_VALUE, max_len=MAX_SEQUENCE_LENGTH):
        # Create a tensor filled with the padding value
        padding_tensor = pt.full((max_len, 1), padding_value)

        # Convert characters in the text to their corresponding indices in the ALPHABET
        value_tensor = pt.LongTensor(
            [[idx for idx, char in enumerate(ALPHABET) if char == t] for t in text]
        )

        # Combine the padding tensor and value tensor, then format the output
        return pad_sequence([padding_tensor, value_tensor])[:, -1, :].permute(1, 0)

    def decode(ids, padding_value=PADDING_VALUE, max_len=MAX_SEQUENCE_LENGTH):
        # Convert indices back to their corresponding characters in the ALPHABET
        return "".join([ALPHABET[idx] for idx in ids.flatten()])

    # Example usage
    encoded_ids = encode("hey ananiya")
    decoded_text = decode(encoded_ids)

    # Print results
    print("Encoded IDs:", encoded_ids)
    print("Decoded Text:", decoded_text)
    import numpy as np
    import torch as pt

    def split(tokens, method=1):
        # Flatten the input tensor
        tokens = tokens.flatten()

        # Determine the splitting method based on the provided method argument
        if method == 1:
            splits = np.split(tokens, np.where(tokens == 1)[0])
        else:
            splits = np.split(tokens, np.where(tokens != -1)[0])

        # Pad and format each split segment
        return [
            pt.nn.functional.pad(
                segment[segment != 0],
                (0, MAX_SEQUENCE_LENGTH - segment[segment != 0].size(0))
            ).unsqueeze(-1).permute(1, 0)
            for segment in splits
            if not pt.equal(segment, pt.tensor([0]))
        ]
    print([_.shape for _ in split(encode("hey hi how are you"), 2)])
    print([_.shape for _ in split(encode("hey hi"))])

    train_labels = []
    train_inputs = []

    # Process the first two examples from the filtered texts
    for text in filtered_texts[:10]:
        # Generate raw input-output pairs by slicing the text
        raw_data = [[text[:idx], char] for idx, char in enumerate(text) if idx > 0]

        # Extract inputs (raw_x) and labels (raw_y)
        raw_inputs = [entry[0] for entry in raw_data]
        raw_labels = [entry[1] for entry in raw_data]

        # Encode the labels (y) and add them to the training labels list
        encoded_labels = pt.Tensor([encode(label, padding_value=0, max_len=1) for label in raw_labels]).unsqueeze(-1)
        train_labels.append(encoded_labels)

        # Encode the inputs (x) and add them to the training inputs list
        encoded_inputs = pt.Tensor(np.array([encode(input_text, padding_value=0, max_len=MAX_SEQUENCE_LENGTH).int() for input_text in raw_inputs])).int()[:, 0, :]
        train_inputs.append(encoded_inputs)

    # Print the shapes of the training labels
    print([entry.shape for entry in train_labels])

    # Stack the inputs and labels into tensors
    train_inputs = pt.vstack(train_inputs)
    print("train_inputs shape:", train_inputs.shape, train_inputs.dtype)
    plt.imshow(train_inputs[:60
               ])

    train_labels = pt.vstack(train_labels)
    print("train_labels shape:", train_labels.shape, train_labels.dtype)
    plt.show()
    plt.imshow(train_labels[:6])

    train_labels = []
    train_inputs = []

    # Process the first two examples from the filtered texts
    for text in filtered_texts[:2]:
        # Generate raw input-output pairs by slicing the text
        raw_data = [[text[:idx], char] for idx, char in enumerate(text) if idx > 0]

        # Extract inputs (raw_x) and labels (raw_y)
        raw_inputs = [entry[0] for entry in raw_data]
        raw_labels = [entry[1] for entry in raw_data]

        # Encode the labels (y) and add them to the training labels list
        encoded_labels = pt.Tensor([encode(label, padding_value=0, max_len=1) for label in raw_labels]).unsqueeze(-1)
        train_labels.append(encoded_labels)

        # Encode the inputs (x) and add them to the training inputs list
        encoded_inputs = pt.Tensor(np.array([encode(input_text, padding_value=0, max_len=MAX_SEQUENCE_LENGTH).int() for input_text in raw_inputs])).int()[:, 0, :]
        train_inputs.append(encoded_inputs)

    # Print the shapes of the training labels
    print([entry.shape for entry in train_labels])

    # Stack the inputs and labels into tensors
    train_inputs = pt.vstack(train_inputs)
    print("train_inputs shape:", train_inputs.shape, train_inputs.dtype)
    plt.imshow(train_inputs[:60
               ])

    train_labels = pt.vstack(train_labels)
    print("train_labels shape:", train_labels.shape, train_labels.dtype)
    plt.show()
    plt.imshow(train_labels[:6])

    import torch
    import torch.nn as nn


    class RotaryPositionalEmbedding(nn.Module):
        def __init__(self, d_model, max_seq_len=1):
            super(RotaryPositionalEmbedding, self).__init__()

            # Create a rotation matrix.
            self.rotation_matrix = torch.zeros(d_model, d_model, device=torch.device(device))
            for i in range(d_model):
                for j in range(d_model):
                    self.rotation_matrix[i, j] = torch.cos(torch.tensor(i) * torch.tensor(j) * torch.tensor(0.01))

            # Create a positional embedding matrix.
            self.positional_embedding = torch.zeros(max_seq_len, d_model, device=torch.device(device))
            for i in range(max_seq_len):
                for j in range(d_model):
                    self.positional_embedding[i, j] = torch.cos(torch.tensor(i) * torch.tensor(j) * torch.tensor(0.01))

        def forward(self, x):
            """
            Args:
                x: A tensor of shape (batch_size, seq_len, d_model).

            Returns:
                A tensor of shape (batch_size, seq_len, d_model).
            """

            # Add the positional embedding to the input tensor.
            x += self.positional_embedding

            # Apply the rotation matrix to the input tensor.
            x = torch.matmul(x, self.rotation_matrix)

            return x
    class CharRoEmbed(nn.Module):
        def __init__(
                self,
                char_vocab: int,
                char_embed_dim: int,
                dropout: float,
                word_embed_dim: int,
                device: torch.device
            ) -> None:
            super(CharRoEmbed, self).__init__()

            # Embedding layer for characters
            self.char_embed = nn.Embedding(char_vocab, char_embed_dim)

            # Positional embedding layer (RotaryPositionalEmbedding, assuming you have defined this)
            self.ro_embed = RotaryPositionalEmbedding(char_embed_dim)

            # Multihead attention layer, with an attention dimension of char_embed_dim // 8
            self.MHA = nn.MultiheadAttention(char_embed_dim, num_heads=char_embed_dim // 8)

            # LSTM layer, processing char_embed_dim into itself
            self.lstm = nn.LSTM(char_embed_dim, char_embed_dim, 1)

            # Flatten the output from LSTM or attention to use it in a linear layer
            self.flatten = nn.Flatten()

            # Dropout to prevent overfitting
            self.dropout = nn.Dropout(dropout)

            # Final linear layer to map from the character embedding space to the word embedding space
            self.linear = nn.Linear(char_embed_dim * MAX_SEQUENCE_LENGTH, word_embed_dim)

            # Assign the device for the model
            self.to(device)

        def forward(self, x: torch.Tensor) -> torch.Tensor:
            # Pass the input through character embedding layer
            x = self.char_embed(x)

            # Optionally apply Rotary Positional Encoding (commented out for now)
            # x = self.ro_embed(x)

            # Optionally apply Multihead Attention (commented out for now)
            # x, _ = self.MHA(x, x, x)

            # Optionally apply LSTM (commented out for now)
            # x, _ = self.lstm(x)

            # Using scaled dot-product attention directly here, as an alternative to LSTM or MHA
            # x = torch.nn.functional.scaled_dot_product_attention(x, x, x, is_causal=True)

            # Flatten the tensor to make it compatible with the linear layer
            x = self.flatten(x)

            # Apply dropout
            x = self.dropout(x)

            # Finally, project to the word embedding space
            x = self.linear(x)

            return x

    import torch
    from torcheval.metrics import Metric
    import Levenshtein as lev

    class EditDistance(Metric):
        def __init__(self, substitution_cost=1, reduction='mean'):
            super().__init__()
            self.substitution_cost = substitution_cost
            self.reduction = reduction
            self._add_state("total_distance", default=torch.tensor(0, dtype=torch.float))  # Use float for total distance
            self._add_state("total_count", default=torch.tensor(0, dtype=torch.int))
            self._add_state("last_distance", default=torch.tensor(0, dtype=torch.float))  # Use float for last distance

        def update(self, preds: list, targets: list):
            """
            Update the metric with predictions and targets.

            Args:
                preds (list of str): List of predicted strings.
                targets (list of str): List of target strings.
            """
            for pred, target in zip(preds, targets):
                distance = lev.distance(pred, target)
                self.total_distance += float(distance)  # Ensure total_distance is treated as float
                self.total_count += len(target)  # Count based on target length
                self.last_distance.copy_(torch.tensor(float(distance), dtype=torch.float))  # Update last distance as float tensor

        def compute(self):
            """Compute the average edit distance."""
            if self.total_count == 0:
                return float('inf')  # Avoid division by zero
            if self.reduction == 'mean':
                return self.total_distance / self.total_count.float()  # Average as float
            return self.total_distance  # Return total distance if no reduction

        def reset(self):
            """Reset the metric for a new evaluation."""
            self.total_distance.zero_()
            self.total_count.zero_()
            self.last_distance.zero_()  # Reset last distance

        def merge_state(self, metrics):
            """Merge state from other metrics."""
            for metric in metrics:
                self.total_distance += metric.total_distance
                self.total_count += metric.total_count
                # Optionally keep track of the last distances from merged metrics
                if metric.last_distance > self.last_distance:
                    self.last_distance.copy_(metric.last_distance)  # Update if the merged metric has a larger last distance

    # Example usage
    metric = EditDistance()
    metric.update(["heF"], ["hey"])
    metric.update(["heiii"], ["hey"])
    average_distance = metric.compute()
    last_distance = metric.last_distance.item()  # Convert tensor to Python number

    print(f"Average Edit Distance: {average_distance}")
    print(f"Last Edit Distance: {last_distance}")

    from torcheval.metrics import MulticlassAccuracy
    from torcheval.metrics import MulticlassPrecision
    from torch.optim.lr_scheduler import ExponentialLR
    from transformers import get_polynomial_decay_schedule_with_warmup

    EMBEDDING_DIM = 32


    class LanguageModel(torch.nn.Module):
        def __init__(self, input_shape, output_shape, dropout_rate, hidden_dim, embedding_dim, num_layers, num_heads, char_embedding_dim, split_method, device):
            super().__init__()
            self.temp_param = nn.Parameter(torch.tensor(1.))
            self.flatten = nn.Flatten()
            self.input_shape = input_shape
            self.dropout_rate = dropout_rate
            self.hidden_dim = hidden_dim
            self.embedding_dim = embedding_dim
            self.num_layers = num_layers
            self.char_embedding_dim = char_embedding_dim

            self.split_method = split_method
            self.char_embedding = CharRoEmbed(len(ALPHABET), self.char_embedding_dim, self.dropout_rate, self.embedding_dim, device)
            self.ff1 = nn.Linear(input_shape, input_shape)
            self.transformer_decoder = nn.TransformerDecoder(nn.TransformerDecoderLayer(input_shape * self.embedding_dim, num_heads, batch_first=True), self.num_layers)
            self.aux_fc = nn.Linear(input_shape * self.embedding_dim, len(ALPHABET))
            self.ff2 = nn.Linear(input_shape * self.embedding_dim, self.hidden_dim)
            self.ff3 = nn.Linear(self.hidden_dim, len(ALPHABET))
            self.to(device)

        def forward(self, input_tensor, log=False):
            if log:
                print("Forward pass")

            input_tensor = input_tensor.unsqueeze(0).permute(1, 0, 2)
            if log:
                print(input_tensor.shape)

            embedded_input = [torch.vstack([self.char_embedding(text.to(device)) for text in split(split_text.cpu())]) for split_text in input_tensor]

            # embedded_input = [torch.vstack([self.char_embedding(text.to(device)) for text in split(split_text.cpu(), torch.round(self.temp_param.cpu()).int())]) for split_text in input_tensor]
            if log:
                print("Embedded text:", len(embedded_input), [tensor.shape for tensor in embedded_input][:20])

            # Padding and stacking the input tensors
            padded_input = [pt.vstack([text, pt.zeros((self.input_shape - text.shape[0]), text.shape[1]).to(device)]) for text in embedded_input]
            if log:
                print("Padded input:", [text.shape for text in padded_input][:20])

            input_tensor = pt.stack(padded_input, dim=0)
            if log:
                print("Stacked tensor:", input_tensor.shape)

            input_tensor = input_tensor.reshape(input_tensor.shape[0], -1)
            if log:
                print("Reshaped tensor:", input_tensor.shape)

            output_shape = input_tensor.shape[0]
            subsequent_mask = torch.tril(torch.ones(output_shape, output_shape)).float().to(device)
            transformer_output = F.relu(self.transformer_decoder(input_tensor, input_tensor, subsequent_mask, subsequent_mask))

            auxiliary_output = transformer_output
            if log:
                print("Transformer output:", transformer_output.shape)

            flattened_output = self.flatten(transformer_output)
            auxiliary_output = self.flatten(auxiliary_output)
            if log:
                print(flattened_output.shape)

            processed_output = F.relu(self.ff2(flattened_output))
            auxiliary_output = self.aux_fc(auxiliary_output)
            final_output = self.ff3(processed_output)

            return final_output, auxiliary_output

        def _round_temp_param(self):
            with torch.no_grad():
                self.temp_param.copy_(torch.round(self.temp_param))

    DEVICE = xm.xla_device()
    device =DEVICE
    run = wandb.init(config=config)
    global model
    if 'model' in locals() or 'model' in globals():
        print('Deleting model')
        del model
    torch.cuda.empty_cache()

    model = LanguageModel(input_shape=200, output_shape=MAX_SEQUENCE_LENGTH, dropout_rate=wandb.config.dropout, hidden_dim=wandb.config.fc_dim,
                          embedding_dim=wandb.config.embed_dim, num_layers=wandb.config.num_layers, num_heads=wandb.config.num_heads,
                          char_embedding_dim=wandb.config.char_embed_dim, split_method=wandb.config.split_method, device=device)

    print(model)
    #model = torch.nn.DataParallel(model)
    model.train().to(device)

    text_table = wandb.Table(columns=["run", "epoch", "loss", "text"])
    optimizer = torch.optim.Adam(model.parameters(), lr=wandb.config.learning_rate)
    # optimizer = torch.amp.syncfree.Adam(model.parameters(), lr=wandb.config.learning_rate)

    # temp_param_optimizer = torch.optim.Adam([model.temp_param], lr=1)
    # scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=wandb.config.scheduler_tmax, eta_min=wandb.config.learning_rate * 0.1)
    losses = []
    edit_distances = []

    train_dataset = TensorDataset(train_inputs, train_labels)
    train_sampler = torch.utils.data.distributed.DistributedSampler(
    train_dataset,
    num_replicas=xm.xrt_world_size(),
    rank=xm.get_ordinal(),
    shuffle=True
    )
    train_loader = DataLoader(
    train_dataset,
    batch_size=wandb.config.batch_size,
    sampler=train_sampler
    )
    #train_loader = DataLoader(train_dataset, batch_size=wandb.config.batch_size, num_workers=4,
    # pin_memory=True)
    criterion = nn.CrossEntropyLoss(ignore_index=0)
    accuracy_metric = MulticlassAccuracy(num_classes=len(ALPHABET), device=device)
    precision_metric = MulticlassPrecision(num_classes=len(ALPHABET), device=device)
    ed_metric = EditDistance()

    total_train_steps = epoch * wandb.config.batch_size
    run_steps = 0

    for epoch_num in range(epoch):
        step_num = 1
        run_steps += 1

        for input_batch, label_batch in train_loader:
            step_num += 1
            # temp_param_optimizer.zero_grad()
            optimizer.zero_grad()


            if wandb.config.auxilary_present == 0:
                output, auxiliary_output = model(input_batch.long().to(device), log=False)
                loss = criterion(output.to(device), label_batch.squeeze(-1).long().to(device))
            elif wandb.config.auxilary_present == 1:
                output, auxiliary_output = model(input_batch.long().to(device), log=False)
                loss = criterion(output.to(device), label_batch.squeeze(-1).long().to(device)) + criterion(auxiliary_output.to(device), label_batch.squeeze(-1).long().to(device))
            elif wandb.config.auxilary_present == -1:
                _, output = model(input_batch.long().to(device), log=False)
                loss = criterion(output.to(device), label_batch.squeeze(-1).long().to(device))

            accuracy_metric.update(output.to(device), label_batch.squeeze(-1).to(device).long())
            precision_metric.update(output.to(device), label_batch.squeeze(-1).to(device).long())
            loss.backward()
            # optimizer.step()
            xm.optimizer_step(optimizer)
            # temp_param_optimizer.step()

        print(epoch_num, scheduler.get_last_lr(), torch.floor(7 * torch.log(0.0025 * torch.tensor(epoch_num - 650) + 1) + 1), torch.floor(7 * torch.log(0.0025 * torch.tensor(epoch_num) + 1) + 1))

        if run_steps == 650:
            run_steps = 0

        model.temp_param = nn.Parameter(torch.floor(7 * torch.log(0.0025 * torch.tensor(run_steps) + 1) + 1)) if run_steps < 651 else nn.Parameter(torch.floor(7 * torch.log(0.0025 * torch.tensor(epoch_num - 650) + 1) + 1))

        text = 'one day a lit'
        for _ in range(50):
            with torch.no_grad():
                if _ > 48:
                    print(text, run.name)
                ed_metric.update([text], [filtered_texts[0]])
                # edit_distances.append(editDistScore(text, rs[0]))
                text = text + decode(F.softmax(model.eval()(encode(text).to(device), False)[0], dim=1).argmax(1))

        text = 'one da'
        for _ in range(60):
            with torch.no_grad():
                run_name = run.name if run.name is not None else "Test"
                if _ > 58:
                    # print(text, run.name)
                    text_table.add_data(run_name, epoch_num, loss.item(), text)
                text = text + decode(F.softmax(model(encode(text).to(device), False)[0], dim=1).argmax(1))
        scheduler.step()
        wandb.log({
            "loss": loss.item(),
            "mean_loss": torch.Tensor(losses).mean().item(),
            "accuracy": accuracy_metric.compute().item(),
            "precision": precision_metric.compute().item(),
            "edit_distance": ed_metric.compute().item(),
            "last_edit_distance": last_distance if isinstance(last_distance, int) or isinstance(last_distance, float) else last_distance.item(),
        })

        losses.append(loss.item())
        plt.plot(losses)

    run.log({"training_samples": text_table})
    accuracy_metric.reset()
    ed_metric.reset()

# Example configuration

config = {
    'dropout': 0.3,
    'fc_dim': 256,
    'embed_dim': 8,
    'num_layers': 8,
    'num_heads': 16,
    'char_embed_dim': 8,
    'learning_rate': 0.00015,
    'batch_size': 512,
    'split_method': 1,
    'auxilary_present': 1,
    "scheduler_tmax":4000,
}
if __name__ == "__main__":
  xmp.spawn(train_model, args=(config,), nprocs=1, start_method='spawn')
# train_model(config)
#wandb.agent(sweep_id, train_model, project="Character-Level Text genration", count=770)

#Wandb.free()
```
